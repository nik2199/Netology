```python
# Среднее отклонение
for column in df.columns:
  if df[column].dtype in ['int64', 'float64']:
      std_value = df[column].std()
      print(f"Среднее отклонение {column} = {std_value}")
```

**3. Рассчитайте и визуализировать корреляционную матрицу для
количественных переменных.
Определите две самые скоррелированные и две наименее
скоррелированные переменные.**

```python
# Корреляционная матрица
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Визуализация
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Корреляционная матрица')
plt.show()
```

```python
# Две самые скоррелированные переменные
max_corr = correlation_matrix.abs().stack().sort_values(ascending=False)
most_correlated = max_corr[max_corr != 1].idxmax()
print("Две самые скоррелированные переменные:", most_correlated)
```

```python
# Две наименее скоррелированные переменные
min_corr = correlation_matrix.abs().stack().sort_values(ascending=True)
least_correlated = min_corr[min_corr != 1].idxmin()
print("По визуализации можно заметить, что наименее скоррелированных переменных несколько, например:")
print("number_project и Work_accident или average_montly_hours и promotion_last_5years")
print("Две наименее скоррелированные переменные по формуле:", least_correlated)
```

**7.Проверить гипотезу, что сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом**

```python
from scipy import stats
```

```python
# Разделим данные на две группы: высокий и низкий оклад
high_salary = df[df['salary'] == 'high']['average_montly_hours']
low_salary = df[df['salary'] == 'low']['average_montly_hours']
# Проводим t-тест
t_statistic, p_value = stats.ttest_ind(high_salary, low_salary)
# Выводим результаты
print(f"T-статистика: {t_statistic}")
print(f"P-значение: {p_value}")
# Проверьте статистическую значимость
alpha = 0.05
print()
print("Вывод:")
if p_value < alpha:
    print("Отвергаем нулевую гипотезу: существует статистически значимая разница во времени, проведенном на работе.")
else:
    print("Принимаем нулевую гипотезу: нет статистически значимой разницы во времени, проведенном на работе.")
```

**8.Рассчитать следующие показатели среди уволившихся и не
уволившихся сотрудников (по отдельности):**

**● Доля сотрудников с повышением за последние 5 лет**

**● Средняя степень удовлетворенности**

**● Среднее количество проектов**

```python
# Показатели для уволившихся
left_df = df[df['left'] == 1]
count_left_promotion = (df[df['left'] == 1]['promotion_last_5years'] == 1).count()
share_left_promotion = (count_left_promotion / df['department'].count())
left_satisfaction_mean = left_df['satisfaction_level'].mean()
left_projects_mean = left_df['number_project'].mean()

print(f"Доля сотрудников с повышением за последние 5 лет = {share_left_promotion:.2%}")
print(f"Средняя степень удовлетворенности = {left_satisfaction_mean:.2}")
print(f"Среднее количество проектов = {left_projects_mean:.2}")
```

```python
# Показатели для не уволившихся
not_left_df = df[df['left'] == 0]
count_left_promotion_not = (df[df['left'] == 0]['promotion_last_5years'] == 1).count()
share_not_left_promotion = (count_left_promotion_not / df['department'].count())
not_left_satisfaction_mean = not_left_df['satisfaction_level'].mean()
not_left_projects_mean = not_left_df['number_project'].mean()

print(f"Доля сотрудников с повышением за последние 5 лет = {share_not_left_promotion:.2%}")
print(f"Средняя степень удовлетворенности = {not_left_satisfaction_mean:.2}")
print(f"Среднее количество проектов = {not_left_projects_mean:.2}")
```

**9.Разделить данные на тестовую и обучающую выборки
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary)
Оценить качество модели на тестовой выборки**

```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Исключаем 'department' и 'salary'
x = df.drop(['left', 'department', 'salary'], axis=1)
y = df['left']
```

```python
# Разделим данные на обучающую и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Модель LDA
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)

# Предсказания на тестовой выборке
y_pred = lda.predict(x_test)

# Оценим качество модели
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Выведем результаты
print("Вывод:\n")
print(f"Модель правильно классифицировала {accuracy:.2%} тестовой выборки")
print(f"Матрица ошибок выглядит следующим образом:\n{conf_matrix}\n")
```
